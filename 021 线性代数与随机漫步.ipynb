# 随机漫步代码

```python
position = 0
walk = [position]
steps = 1000  # 累积走的位置数量
expr = 10 # 要求的到达距离
for i in range(steps):
    step = 1 if np.random.randint(0,2) else -1 # 随机前进或者后退一步 1，-1
    position += step # 累积步长
    walk.append(position) 
plt.figure(figsize=(16,9))
plt.plot(walk)
plt.plot(np.full(steps,10))
(np.abs(walk)>expr).argmax()
```

# 有几次到达了指定位置

```python
(np.array(walk)==expr).sum()
```

# 到达指定位置的步数分别是

```python
np.argwhere(np.array(walk)==expr)
```
